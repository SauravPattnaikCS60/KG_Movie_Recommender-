```python
import ast

def find_unique_dept(l_json):
    dept_list = []
    for json in l_json:
        if 'department' in json.keys():
            dept_list.append(json['department'])
    
    return list(set(dept_list))

credits['departments'] = credits['crew'].apply(ast.literal_eval).apply(find_unique_dept)
```

```python
def return_film_cast(cast_json):
    cast_json = ast.literal_eval(cast_json)
    cast_members = []
    for json in cast_json:
        if 'name' in json.keys():
            cast_members.append(json['name'])
    return cast_members

credits['cast_members'] = credits['cast'].apply(return_film_cast)
credits['cast_members'].head()
```

```python
def return_film_creative_team(crew_json):
    crew_json = ast.literal_eval(crew_json)
    creative_filters = ['Directing','Writing','Editing']
    creative_members = []
    for json in crew_json:
        if 'department' in json.keys():
            dept = json['department']
            if dept in creative_filters:
                if 'name' in json.keys():
                    creative_members.append(json['name'])
    return creative_members

credits['creative_members'] = credits['crew'].apply(return_film_creative_team)
credits['creative_members'].head()
```

## FILM PLOT ANALYSIS

```python
keywords = pd.read_csv('../data/keywords.csv')
keywords.head()
```

```python
keywords['keywords'].iloc[5]
```

```python
def return_film_plot_keywords(plot_json):
    plot_json = ast.literal_eval(plot_json)
    plot_keywords = []
    for json in plot_json:
        if 'name' in json.keys():
            plot_keywords.append(json['name'])
    
    return plot_keywords

keywords['plot_keywords'] = keywords['keywords'].apply(return_film_plot_keywords)
keywords['plot_keywords'].head()
```
